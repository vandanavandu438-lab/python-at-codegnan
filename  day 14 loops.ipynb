```python
s=input()
m=s.replace(" ","")
vowels=["a","e","i","o","u"]
vowelcount=0
consonantcount=0
for ch in m:
    if ch in vowels:
        vowelcount += 1
for ch in m:
    if ch.isalpha():
        if ch not in vowels:
            consonantcount += 1
print(vowelcount)
print(consonantcount)
        
```

```python
s=input()
m=s.replace(" ","")
vowels="aeiou"
vowelcount=0
consonantcount=0
i=0
while i<len(m):
    if m[i] in vowels:
        vowelcount += 1
    i += 1
j=0
while j<len(m):
    if m[j].isalpha():
        if m[j] not in vowels:
            consonantcount += 1
    j += 1
print(vowelcount)
print(consonantcount)
```
