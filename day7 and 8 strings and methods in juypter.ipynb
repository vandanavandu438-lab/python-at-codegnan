```python
#string methods
-case conversion methods
 -lower()-it converts entire string into lower case
 -upper()-it converts entire string into upper case
 -title()-it converts first char of every word into upper case remaining into lower case in a string 
 -swapcase()- it converts lower case char into upper case and upper case into lowercase in a string
 -captilize()- it captilizes the fisrt char of the statement into upper case and remaining into lower case in a string
```

```python
#white space or pattern removing
-strip()-it removes the white space to both side of string
-lstrip()-it removes the left side white space to the string
-rstrip()-it removes the right side white space to the string
```

```python
s="       codegnan       "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s="@@@@@@code@gnan@@@@@"
print(s)
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip("@"))
```

```python
#count the frequency of the sub string
-count(sub_string)-it returns the total count of non overlapping substring
```

```python
s="ABCACAC"
print(s.count("C"))
print(s.count("CAC"))
```

```python
s="ABCACACcccc"
print(s.count('c')+s.count('C'))
s1=s.lower()
print(s1.count('c'))
```

```python
#split()-it splits the entire string intlo list of substring,bydefault it splits at space
s="i am a python batch student"
print(s.split())
print(s.split("a"))
print(s.split("t"))
```

```python
#split()-it splits the entire string intlo list of substring,bydefault it splits at space
s="i am a pyThon batch student"
print(s.lower().split("t"))
```

```python
#"sep".join(iterable)-it combines the all items in iterable with separator
words=['i', 'am', 'a', 'python', 'batch', 'student']
print("".join(words))
print(" ".join(words))
print("_".join(words))
print("-".join(words))
```

```python
#replace(old,new)-it replace the old string with new string
s="python"
print(s.replace("p","j"))
print("AABBAAABBAABBCC".replace("A","Z",5)) #it replaces at upto 5 places
print("AABBAAABBAABBCC".replace("A","Z",100))
```

```python
#string checking methods
 -islower()-it returns true if entire string is in lower case 
 -isupper()-it returns true if entire string is in upper case
 -title()-it converts first char of every word into upper case remaining into lower case in a string 
 -swapcase()- it converts lower case char into upper case and upper case into lowercase in a string
 -captilize()- it captilizes the fisrt char of the statement into upper case and remaining into lower case in a string
 -isdigit()-it returns true if entire string contains all digits
 -isalpha()-it returns true if entire string conatins alphabets
 -isalnum()-it return true if entire string conatins either alphabets or digits
 -startswith(sub_string)-it returns true if string starts with substring 
 -endswith(sub_string)-it returns true if string ends with substring 
```

```python
print("code".islower())
print("Code".islower())
print("code123".islower())
print("CODE123".isupper())
print("1234a".isalnum())
print("codegnan".startswith("co"))
```
