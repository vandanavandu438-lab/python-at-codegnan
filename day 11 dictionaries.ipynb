```python
#dictionary methods
-value getting from dictionary
 -get(key,default_value)-it returns specified key's value if it exists in dictionary otherwise default value
-add or update dictionary items
 - upadte(new_dict)-it upadtes dictionary items with new_dict if keys already exists otherwise,items from dictionary
-get keys,values,items from dictionary
 -keys()-it returns list of dictionary keys
 -values()-it return list of dictionary values
 -items()-it returns list of tuple of key-value pairs
-delete items from dictionary
 -pop(key)-it deletes the items and returns the value of specified key,otherwise keyerror
 -popitem()-it defines and returns the last item from the dictionary
```

```python
#keys(),value(),items()
details={'name':'vandana','age':30,'course':'python'}
print(details.keys())
print(details.values())
print(details.items())
print(type(details))
```
