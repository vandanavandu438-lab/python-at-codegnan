```python
num=int(input())
if num > 0:
    print("positive")
elif num < 0:
    print("negative")
else:
    print("zero")
```

```python
num1=int(input())
num2=int(input())
print(f"their sum: {num1+num2}")
print(f"their difference: {num1-num2}")
print(f"their product: {num1*num2}")
if num1==0 or num2==0:
    print("error")
else:
    print(f"their quotient: {num1/num2}")
print(f"their remainder: {num1%num2}")
```

```python
amount=int(input())
if amount < 500:
    print("no discount")
elif 500 <= amount <= 999:
    print("5% discount")
elif 1000 <= amount <= 1999:
    print("10% discount")
else:
    print("20 discount")
```

```python
sentence=input()
word=input()
if word in sentence:
    print(sentence.replace(word,word.upper()))
else:
    print(sentence.lower())
```

```python
n=int(input())
scores=list(map(int,input().split()))
if n>5:
    scores.sort(reverse=True)
    print(scores[:3])
else:
    print("Not enough scores")
```

```python
d={}
d["student_name"]=input()
d["marks"]=int(input())
d["Grade"]=input()
if d["marks"]>=70:
    d.update({"status":"pass","scholarship":"True"})
    print(f"{d["student_name"]} -> {d["status"]}")
    print(f"keys: {list(d)}")
else:
    print("No update required")
```

```python
sales=tuple(map(int,input().split()))
if max(sales)>=50000:
    print(f"Highest Sales:{max(sales)}")
    print(f"Lowest Sales:{min(sales)}")
    print(f"Average Sales:{sum(sales)/6:.2f}")
else:
    print("Sales target not achieved")
```
